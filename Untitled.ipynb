```python
df['action_date'] = pd.to_datetime(df['action_date'])
df['creation_date'] = pd.to_datetime(df['creation_date'])
df['view_date'] = pd.to_datetime(df['view_date'])
```

```python

# Calculate time differences
df['creation_to_view_duration'] = df['view_date'] - df['creation_date']
df['creation_to_action_duration'] = df['action_date'] - df['creation_date']

df['creation_to_view_duration_hours'] = (df['view_date'] - df['creation_date']).dt.total_seconds() / 3600
df['creation_to_action_duration_hours'] = (df['action_date'] - df['creation_date']).dt.total_seconds() / 3600
df['action_to_view_duration_hours'] = (df['action_date'] - df['creation_date']).dt.total_seconds() / 3600
```

```python
duration_performance = df[['user_team','user_actioned','creation_to_action_duration_hours','customer_problem_resolved']]
duration_performance
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'average_durations_per_user' is your DataFrame
plt.figure(figsize=(12, 6))

# Combine the two duration columns into a single 'duration' column
df_long = pd.melt(average_durations_per_user, id_vars=['user_team'], value_vars=['creation_to_action_duration_hours', 'creation_to_view_duration_hours'])

# Violin plot with 'hue' set to the variable column
ax = sns.violinplot(x='user_team', y='value', hue='variable', data=df_long, split=True, inner='quartile', palette="Set3")

# Calculate and plot means
means = df_long.groupby(['user_team', 'variable'])['value'].mean().reset_index()

for i, team in enumerate(df_long['user_team'].unique()):
    for j, variable in enumerate(df_long['variable'].unique()):
        mean_value = means[(means['user_team'] == team) & (means['variable'] == variable)]['value'].values[0]
        ax.scatter(i + j * 0.2 - 0.1, mean_value, color='black', marker='o', s=100, label=f'{variable} Mean' if j == 0 else '')

plt.title('Violin Plot of Creation Durations for Teams with Means')
plt.xlabel('User Team')
plt.ylabel('Duration (hours)')
plt.legend(title='Duration Type')
plt.show()
```
