```python
df_employee = df.groupby(['user_team', 'user_actioned', 'customer_problem_resolved']).count().reset_index()
df_employee
```

```python
solved_to_total_r = []
for i in range(0, len(df_employee), 2):
    s_t_r = df_employee.iloc[i+1, 3] / (df_employee.iloc[i+1, 3] + df_employee.iloc[i, 3]) * 100
    solved_to_total_r.append(s_t_r)

solved_to_total_r

df_employee_status = df.groupby(['user_team', 'user_actioned']).count()
df_employee_status = df_employee_status.reset_index()
df_employee_status['Solved to total percent'] = solved_to_total_r
df_employee_status
```

```python
selected_col = ['user_team', 'user_actioned', 'Solved to total percent']
employee_team_effeciency = df_employee_status[selected_col].copy()
employee_team_effeciency
```

```python
import matplotlib.pyplot as plt

# Sample data
user_teams = ['blue', 'green', 'orange', 'red']
solved_percent = [43.649965, 57.072886, 27.174133, 73.716962]

# Sort the data in descending order
sorted_data = sorted(zip(user_teams, solved_percent), key=lambda x: x[1], reverse=True)
sorted_teams, sorted_percent = zip(*sorted_data)

# Create a horizontal bar graph
plt.figure(figsize=(10, 6))
plt.barh(sorted_teams, sorted_percent, color=sorted_teams)
plt.xlabel('Solved to Total Percent')
plt.ylabel('User Team')
plt.title('Solved to Total Percent by User Team')
plt.gca().invert_yaxis()  # Invert y-axis to have the highest bar at the top
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
employee_efficiencies = {
    'orange': [25.265188, 30.470914, 28.234183, 26.372549, 25.527831],
    'blue': [31.950207, 30.294906, 66.333938, 40.489130, 58.691589, 34.140017],
    'green': [66.126280, 57.213115, 70.542636, 34.409516],
    'red': [77.858338, 69.288606, 72.467823, 77.603583, 71.366460],
}

# Get the number of employees in each team
team_sizes = {team: len(employees) for team, employees in employee_efficiencies.items()}

# Create a scatter plot
plt.figure(figsize=(10, 6))
for team, efficiencies in employee_efficiencies.items():
    plt.scatter(efficiencies, [team]*len(efficiencies), s=team_sizes[team]*100, alpha=0.5, label=team)

plt.xlabel('Efficiency (%)')
plt.ylabel('User Team')
plt.title('Employee Efficiencies with Team Sizes')
plt.grid(True)
plt.show()
```

```python
team_effeciency = employee_team_effeciency.groupby('user_team')['Solved to total percent'].mean().reset_index()
team_effeciency
```

```python
import circlify


data_dict = []
for index, row in team_effeciency.iterrows():
    child_array = []
    cat_dict = dict(id=team_effeciency.iloc[index,0], datum=team_effeciency.iloc[index,1])
    data_dict.append(cat_dict.copy())

circle_df = pd.DataFrame(data_dict).sort_values(by='datum', ascending=False)
# circle_df['color'] = PURPLE
# circle_df.loc[3, 'color'] = BLUE
# circle_df.loc[10, 'color'] = YELLOW

circles = circlify.circlify(
    data_dict, 
    show_enclosure=False, 
    target_enclosure=circlify.Circle(x=0, y=0, r=1)
)


fig, ax = plt.subplots(figsize=(8,8))


ax.axis('off')

lim = max(
    max(
        abs(circle.x) + circle.r,
        abs(circle.y) + circle.r,
    )
    for circle in circles
)
plt.xlim(-lim, lim)
plt.ylim(-lim, lim)

for index, circle in enumerate(circles):
    if circle.level == 1:    
        x, y, r = circle
        label = circle.ex["id"]
        # edgecolor = BLUE if label == 'Bath' else ( YELLOW if label == 'Bed' else PURPLE)
        ax.add_patch( plt.Circle((x, y), r, linewidth=2, color=label))
        plt.annotate(label, (x,y ), ha='center', color="white")

plt.show()
```

```python
df_customer_satisfaction = df.groupby(['ticket_type', 'customer_satisfaction']).count().reset_index()
df_customer_satisfaction
```

```python
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 1, 'color'] = '#f0f7da'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 2, 'color'] = '#c9df8a'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 3, 'color'] = '#77ab59'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 4, 'color'] = '#36802d'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 5, 'color'] = '#234d20'
df_customer_satisfaction
```

```python
dict_customer_satisfaction = []

for i in range(0, len(df_customer_satisfaction), 5):
    if(df_customer_satisfaction.iloc[i, 0] == 'email'):
        dict_customer_satisfaction.append(dict(ticket_type='email', unhappy=df_customer_satisfaction.iloc[i, 2]+df_customer_satisfaction.iloc[i+1, 2], average=df_customer_satisfaction.iloc[i+2, 2], happy=df_customer_satisfaction.iloc[i+3, 2]+df_customer_satisfaction.iloc[i+4, 2]))
    if(df_customer_satisfaction.iloc[i, 0] == 'live_chat'):
        dict_customer_satisfaction.append(dict(ticket_type='live_chat', unhappy=df_customer_satisfaction.iloc[i, 2]+df_customer_satisfaction.iloc[i+1, 2], average=df_customer_satisfaction.iloc[i+2, 2], happy=df_customer_satisfaction.iloc[i+3, 2]+df_customer_satisfaction.iloc[i+4, 2]))
    if(df_customer_satisfaction.iloc[i, 0] == 'web_form'):
        dict_customer_satisfaction.append(dict(ticket_type='web_form', unhappy=df_customer_satisfaction.iloc[i, 2]+df_customer_satisfaction.iloc[i+1, 2], average=df_customer_satisfaction.iloc[i+2, 2], happy=df_customer_satisfaction.iloc[i+3, 2]+df_customer_satisfaction.iloc[i+4, 2]))
customer_happiness = pd.DataFrame(dict_customer_satisfaction)
customer_happiness
```

```python
melted_df = pd.melt(customer_happiness, id_vars=['ticket_type'], var_name='satisfaction', value_name='count')

df_customer_happiness = melted_df.to_dict(orient='list')
df_customer_happiness['satisfaction'] = df_customer_happiness.pop('count')

df_customer_happiness = pd.DataFrame(df_customer_happiness)
df_customer_happiness = df_customer_happiness.sort_values(by=['ticket_type'])
df_customer_happiness['color'] =  ['#b3daf6', '#1b8fe4','#09304c','#ffaac1', '#ff0045','#550017','#ffa523', '#ff9700','#553200']
df_customer_happiness['rating'] =  ['Bad','Avg.','Good','Bad','Avg.','Good','Bad','Avg.','Good']
df_customer_happiness
```

```python
df_customer_satisfaction_tt = df.groupby(['ticket_type']).count().reset_index()
df_customer_satisfaction_tt['labels'] = ['Email', 'Live Chat', 'Web Form']
df_customer_satisfaction_tt
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

inner_color = ['#156eaf', '#ff3169', '#ffa523']

ax.pie(df_customer_happiness['satisfaction'], labels=df_customer_happiness['rating'],labeldistance=0.75, colors= df_customer_happiness['color'], wedgeprops=dict(edgecolor='w', alpha=0.7), textprops=dict(fontsize='small',fontweight='bold', color='w'))
ax.pie(df_customer_satisfaction_tt['creation_date'], labels=df_customer_satisfaction_tt['labels'], colors=inner_color,labeldistance=0.4, radius=0.5, wedgeprops=dict(edgecolor='w'), textprops=dict(color='w'))

ax.set_title("Customer Rating to Ticket Type")
plt.show()
```
