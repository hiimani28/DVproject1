```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('tickets.csv')
```

```python
df_employee = df.groupby(['user_team', 'user_actioned', 'customer_problem_resolved']).count().reset_index()
```

```python
solved_to_total_r = []
for i in range(0, len(df_employee), 2):
    s_t_r = df_employee.iloc[i+1, 3] / (df_employee.iloc[i+1, 3] + df_employee.iloc[i, 3]) * 100
    solved_to_total_r.append(s_t_r)

solved_to_total_r

df_employee_status = df.groupby(['user_team', 'user_actioned']).count()
df_employee_status = df_employee_status.reset_index()
df_employee_status['Solved to total percent'] = solved_to_total_r
df_employee_status
```

```python
selected_col = ['user_team', 'user_actioned', 'Solved to total percent']
employee_team_effeciency = df_employee_status[selected_col].copy()
employee_team_effeciency
```

```python
df_customer_satisfaction = df.groupby(['ticket_type', 'customer_satisfaction']).count().reset_index()
df_customer_satisfaction
```

```python
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 1, 'color'] = '#f0f7da'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 2, 'color'] = '#c9df8a'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 3, 'color'] = '#77ab59'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 4, 'color'] = '#36802d'
df_customer_satisfaction.loc[df_customer_satisfaction['customer_satisfaction'] == 5, 'color'] = '#234d20'
df_customer_satisfaction
```

```python
dict_customer_satisfaction = []

for i in range(0, len(df_customer_satisfaction), 5):
    if(df_customer_satisfaction.iloc[i, 0] == 'email'):
        dict_customer_satisfaction.append(dict(ticket_type='email', unhappy=df_customer_satisfaction.iloc[i, 2]+df_customer_satisfaction.iloc[i+1, 2], average=df_customer_satisfaction.iloc[i+2, 2], happy=df_customer_satisfaction.iloc[i+3, 2]+df_customer_satisfaction.iloc[i+4, 2]))
    if(df_customer_satisfaction.iloc[i, 0] == 'live_chat'):
        dict_customer_satisfaction.append(dict(ticket_type='live_chat', unhappy=df_customer_satisfaction.iloc[i, 2]+df_customer_satisfaction.iloc[i+1, 2], average=df_customer_satisfaction.iloc[i+2, 2], happy=df_customer_satisfaction.iloc[i+3, 2]+df_customer_satisfaction.iloc[i+4, 2]))
    if(df_customer_satisfaction.iloc[i, 0] == 'web_form'):
        dict_customer_satisfaction.append(dict(ticket_type='web_form', unhappy=df_customer_satisfaction.iloc[i, 2]+df_customer_satisfaction.iloc[i+1, 2], average=df_customer_satisfaction.iloc[i+2, 2], happy=df_customer_satisfaction.iloc[i+3, 2]+df_customer_satisfaction.iloc[i+4, 2]))
customer_happiness = pd.DataFrame(dict_customer_satisfaction)
customer_happiness
```

```python
melted_df = pd.melt(customer_happiness, id_vars=['ticket_type'], var_name='satisfaction', value_name='count')

df_customer_happiness = melted_df.to_dict(orient='list')
df_customer_happiness['satisfaction'] = df_customer_happiness.pop('count')

df_customer_happiness = pd.DataFrame(df_customer_happiness)
df_customer_happiness = df_customer_happiness.sort_values(by=['ticket_type'])
df_customer_happiness['color'] =  ['#c6e3f8', '#54abeb','#146bab','#ffbfd1', '#ff4074','#bf0034','#ffe5bf', '#ffb140','#bf7100']
df_customer_happiness['rating'] =  ['Bad','Avg.','Good','Bad','Avg.','Good','Bad','Avg.','Good']
df_customer_happiness
```

```python
df_customer_satisfaction_tt = df.groupby(['ticket_type']).count().reset_index()
df_customer_satisfaction_tt['labels'] = ['Email', 'Live Chat', 'Web Form']
df_customer_satisfaction_tt
```

```python
fig, ax = plt.subplots(figsize=(20, 8), facecolor='none')

inner_color = ['#156eaf', '#ff3169', '#ffa523']

ax.pie(df_customer_happiness['satisfaction'], labels=df_customer_happiness['rating'],labeldistance=0.75, colors= df_customer_happiness['color'], wedgeprops=dict(edgecolor='w', alpha=0.8), textprops=dict(fontsize='small',fontweight='bold', color='w'))
ax.pie(df_customer_satisfaction_tt['creation_date'], labels=df_customer_satisfaction_tt['labels'], colors=inner_color,labeldistance=0.4, radius=0.5, wedgeprops=dict(edgecolor='w'), textprops=dict(color='w'))

ax.set_title("Impact of Ticket Type on Ratings", y=0)
plt.show()
```

```python
result = df.groupby(['user_team', 'customer_problem_resolved']).size().reset_index(name='count')

pivot_result = result.pivot(index='user_team', columns='customer_problem_resolved', values='count').reset_index()
pivot_result.columns.name = None
pivot_result.columns = ['user_team', 'False', 'True']
result_dict = pivot_result.to_dict(orient='list')

df_new = pd.DataFrame(result_dict)
df_new.set_index('user_team', inplace=True)
df_new['Total'] = df_new['False']+df_new['True']
# Sort by the "True" column in descending order
df_new.sort_values(by='Total', ascending=True, inplace=True)
df_new.drop(columns='Total', inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(8, 4), facecolor='none')
df_new.plot(kind='barh', stacked=True, color=['#FF3169', '#00B4B0'], width=0.8, ax=ax, edgecolor='w')

ax.set_ylabel('')
ax.set_yticklabels(['Green', 'Orange', 'Blue', 'Red'])
ax.legend(labels=['Unresolved', 'Resolved'], bbox_to_anchor=(0.5, 1.08), loc='upper center',ncol=2)

ax.set_title('Breakdown of Ticket Resolutions by Team',y=1.09)

for team in df_new.index:
    total = df_new.loc[team].sum()
    false_percent = df_new.loc[team, 'False'] / total
    true_percent = df_new.loc[team, 'True'] / total

    # Finding x coordinate centers
    false_center = df_new.loc[team, 'False'] / 2
    true_center = df_new.loc[team, 'False'] + df_new.loc[team, 'True'] / 2

    ax.annotate(f'{false_percent:.1%}',
                (false_center, list(df_new.index).index(team)),
                ha='center', va='center', color='white', fontsize=10)
    ax.annotate(f'{true_percent:.1%}',
                (true_center, list(df_new.index).index(team)),
                ha='center', va='center', color='w', fontsize=10)
plt.xticks([])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
plt.savefig('plt2.pdf')
ax.set_facecolor('none')
plt.show()
```

```python
df = df[df['customer_problem_resolved']] # The data only shows where the issue was RESOLVED!!!!!

df['action_date'] = pd.to_datetime(df['action_date'])
df['creation_date'] = pd.to_datetime(df['creation_date'])
df['view_date'] = pd.to_datetime(df['view_date'])

df['creation_to_action_duration'] = df['action_date'] - df['creation_date']
df['creation_to_action_duration_hours'] = (df['action_date'] - df['creation_date']).dt.total_seconds() / 3600 # Represents the time between creating the ticket and closing the ticket - so time taken to resolve a ticket
average_durations_per_user = df.groupby(['user_team', 'user_actioned']).agg({
    'creation_to_action_duration_hours': 'mean'}).reset_index()

data=average_durations_per_user[['user_team','user_actioned', 'creation_to_action_duration_hours']]

avg_hours = data['creation_to_action_duration_hours'].mean()
```

```python
from matplotlib.patches import Rectangle

colors = {'blue': '#66dcff', 'green': '#00B4B0', 'orange': '#ffeacc', 'red': '#FF3169'}
fig, ax = plt.subplots(figsize=(10, 6), facecolor='none')
sns.swarmplot(data=data, x='user_actioned', y='creation_to_action_duration_hours', 
            hue='user_team',s=15, palette=colors, ax=ax)

ax.set_title('Ticket Lifespan Variation Across User Teams (the lower the better)', y=1.1)

ax.legend(bbox_to_anchor=(0.5, 1.08), loc='upper center',ncol=4)
# Draw a rectangle on the plot with equal aspect ratio (a square)
square = Rectangle((14.4, 9.5), 5, 11.7, fc='none',ec='#00ff00', lw=0.5)
ax.add_patch(square)
square = Rectangle((9.5, 42.2), 6, 14, fc='none',ec='#ff0000', lw=0.5)
ax.add_patch(square)

ax.axhline(y=avg_hours, color='k', linestyle='--', alpha=0.5)

ax.set_xticks([])
ax.set_xlabel('User Actioned')
ax.set_ylabel('Time spent on ticket')
ax.set_yticks([10,20,30,avg_hours,40,50])
ax.set_yticklabels(['10 hrs', '20', '30', '31.9 (avg)', '40', '50 hrs'])

ax.spines['top'].set_visible(False)
ax.set_facecolor('none')
ax.spines['right'].set_visible(False)
# Show the plot
plt.show()
```
